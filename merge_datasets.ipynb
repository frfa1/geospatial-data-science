```python
# Imports
import geopandas as gpd
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Read boligzonen dataset
bz = pd.read_csv("data/boligzonen.csv")
bz.head()
```

```python
# Read boligsiden dataset
bs = pd.read_csv("data/boligsiden.csv")
bs.head()
```

```python
# Rename boligsiden columns to match boligzonen naming
bs = bs.rename(columns={"Sagsnr": "id", "per_area_price": "norm_rent" })
bs.head()
```

```python
# Drop boligsiden columns that are not in boligzonen
bs = bs.drop(["available_from", "address", "zip_code"], axis=1)
bs.head()
```

```python
# Order of columns in boligsiden makes more sense, rearrange columns in boligzonen
bz = bz[["id", "rooms", "area", "rent", "norm_rent", "latitude", "longitude"]]
bz.head()
```

```python
# Show number of records for each dataset
f"boligzonen records : {len(bz)}, boligsiden records : {len(bs)}"
```

```python
# Merge datasets and verify that number of records is correct
merged = pd.concat([bz, bs], ignore_index=True)
len(merged) == len(bz) + len(bs)
```

```python
# Save merged dataset
merged.to_csv("data/merged.csv", index=False)
```
